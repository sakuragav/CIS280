```python
import pandas as pd
import seaborn as sns
# i
url = 'https://raw.githubusercontent.com/nguyen-toan/ISLR/master/dataset/College.csv'
df = pd.read_csv(url)
df = df.rename(columns = {'Unnamed: 0':'Name'})
college_new = df[["Name", "Private", "Apps", "Accept", "Enroll", "PhD", "S.F.Ratio", "Grad.Rate"]]
college_new.head()
```

```python
# ii
private = college_new[college_new['Private'] == 'Yes']
public = college_new[college_new['Private'] != 'Yes']
```

```python
# iii : Create a comparative boxplot of Graduation rates of public and private universities.
sns.boxplot(y = 'Grad.Rate', x = 'Private', data = college_new, hue = 'Private')
```

```python
# iii : Create a comparative boxplot of number of “S.F.Ratio”” in public and private universities
sns.boxplot(y = 'S.F.Ratio', x = 'Private', data = college_new, hue = 'Private')
```
