```python
import pandas as pd
import plotly.express as px
df = pd.read_csv("https://raw.githubusercontent.com/covid19india/deep-dive/master/data/dataset/covid_19_india.csv")
fig = px.bar(df, x = 'State/UnionTerritory', y = 'Confirmed')
fig.show()
```

```python
# Finding NaN Value

count_nan_in_df = df.isnull().sum()
print('Count of NaN: ' + str(count_nan_in_df))
```

 Using the appropriate Python function to drop the missing values from the
entire dataset. Make sure to include your code and output.

```python
df = df.dropna()
count_nan_in_df = df.isnull().sum()
print('Count of NaN: ' + str(count_nan_in_df))
```

Using Plotly, create an interactive scatter plot showing the number of confirmed cases in each state by setting the animation frame as the time column. Make
sure to include your code and output.

```python
g = px.scatter(df, x = 'State/UnionTerritory', y = 'Confirmed', color = 'State/UnionTerritory',animation_frame= 'Time')
g.show()
```
