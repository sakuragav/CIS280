```python
import pandas as pd
import seaborn as sns
from numpy import mean, median
reviews = pd.read_csv("https://raw.githubusercontent.com/Nhandsome/for_kaggle/main/0.doc_for_me/data/ign_scores.csv")
reviews_df = pd.DataFrame(reviews)
reviews_df.head()

```

```python
#Create a bar chart that shows the average score for racing games, for each platform.
#Your chart should have one bar for each platform and have the title Average Score for Racing Games, by Platform.
import matplotlib.pyplot as plt

g = sns.barplot(x = "Platform", y = "Racing", data = reviews_df, estimator = mean)
g.set_xticklabels(reviews_df['Platform'], rotation = 90)
g.set_title("Average Score for Racing Games, by Platform")
```

```python
# Rename columns
new_df = pd.melt(reviews_df, id_vars = ['Platform'])  # changes the format where one or more columns are identifiers
# melt function condenses the data - in this example, it changes the columns of all the years into one 'year' column
columns = ['Platform', 'Genre', 'Review']
new_df.columns = columns
new_df.head()
```

```python
#Use the data to create a heatmap of average score by genre and platform. Include ’annot = True’ as an additional input argument.
#Your map should have the title Average Score for Games, by Platform and Genre and x-axis label as Genre.
df_pivoted = new_df.pivot("Platform", "Genre", "Review")
ax = sns.heatmap(df_pivoted, annot = True)
ax.set(xlabel='Genre', ylabel='Platform')
ax.set_title("Average Score for Games, by Platform and Genre")
```
