```python
columns = ['Country', 'Code', 'Year','Percent Obese']
df = pd.read_csv("ObesityData.csv", names = columns, skiprows = 1)
df.head()
```

```python
import plotly.express as px
df2014 = df.query("Year==2014")

fig = px.choropleth(df2014, locations = 'Code',
                    color = "Percent Obese",
                    hover_name = "Country",
                    color_continuous_scale = px.colors.sequential.Plasma)
fig.show()
```
